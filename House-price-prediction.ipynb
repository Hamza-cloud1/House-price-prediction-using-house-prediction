```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
df1=pd.read_csv(r'D:\data sets\housing-prices-1.csv')
df2=pd.read_csv(r'D:\data sets\housing-prices-2.csv')
df3=pd.concat([df1,df2],axis=1)
# df3.info()
#imputation initial
df_mean = df3.copy()
#imputation for normal data
mean_imputer = SimpleImputer(strategy='mean')
df_mean['total_bedrooms'] = mean_imputer.fit_transform(df_mean['total_bedrooms'].values.reshape(-1,1))
#imputation for skewed
mean_imputer = SimpleImputer(strategy='median')
df_mean['bedrooms'] = mean_imputer.fit_transform(df_mean['bedrooms'].values.reshape(-1,1))
df_mean['bathrooms'] = mean_imputer.fit_transform(df_mean['bathrooms'].values.reshape(-1,1))
df_mean['sqft_living'] = mean_imputer.fit_transform(df_mean['sqft_living'].values.reshape(-1,1))
df_mean['sqft_lot'] = mean_imputer.fit_transform(df_mean['sqft_lot'].values.reshape(-1,1))
df_mean['floors'] = mean_imputer.fit_transform(df_mean['floors'].values.reshape(-1,1))
df_mean['price'] = mean_imputer.fit_transform(df_mean['price'].values.reshape(-1,1))
df_mean['condition'] = mean_imputer.fit_transform(df_mean['condition'].values.reshape(-1,1))
df_mean['sqft_above'] = mean_imputer.fit_transform(df_mean['sqft_above'].values.reshape(-1,1))
df_mean['sqft_basement'] = mean_imputer.fit_transform(df_mean['sqft_basement'].values.reshape(-1,1))
df_mean['yr_built'] = mean_imputer.fit_transform(df_mean['yr_built'].values.reshape(-1,1))
df_mean['yr_renovated'] = mean_imputer.fit_transform(df_mean['yr_renovated'].values.reshape(-1,1))
#imputation for categorical data
df_mean['city'] = df_mean['city'].fillna(df_mean['city'].mode()[0])
df_mean['street'] = df_mean['street'].fillna(df_mean['street'].mode()[0])
df_mean['statezip'] = df_mean['statezip'].fillna(df_mean['statezip'].mode()[0])
df_mean['country'] = df_mean['country'].fillna(df_mean['country'].mode()[0])
df_mean.isna().sum()
df=df_mean.copy()
# df.isna().sum()
#choosing parameters for training
area=df['sqft_living']
price=df['price']
x=np.array(area).reshape(-1,1)
y=np.array(price)
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state = 100)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
# print('Feature input-y',y_train)
# print('Prediction-y',y_test)
#Fit the model over the training dataset
model = LinearRegression()
model.fit(X_train,y_train)
# print('Dataset training outcomes:')
# print('intercept',model.intercept_)
# print('coefficient',model.coef_)
pred=model.predict(X_test)
predictions = pred.reshape(-1,1)
#Calculate mean squared error to evaluate model performance
print('mean square error : ', mean_squared_error(y_test,predictions))
#calculate r square error to evaluate model performance
print('R-squared error : ', r2_score(y_test,predictions))
#1
plt.scatter(x,y, color='red')
plt.plot(x,model.predict(X_test),color='blue')
plt.title("linear regression")
plt.xlabel('area')
plt.ylabel('price')
plt.show()
#2
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(x, y)
ax.plot(X_train, y_train)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')


```
